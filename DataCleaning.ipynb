```python
import pandas as pd
```

```python
df = pd.read_csv('DataAnalyst.csv')
```

```python
df
```

```python
df = df[df['Salary Estimate'] != '-1']
```

```python
df
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.split('(')[0])
```

```python
df
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.lower().replace('k', ''))
```

```python
df
```

```python
df['Salary Estimate']=df['Salary Estimate'].apply(lambda x: x.replace('$', ''))
```

```python
df
```

```python
df['Min Salary'] = df['Salary Estimate'].apply(lambda x: x.split('-')[0])
```

```python
df
```

```python
df['Max Salary'] = df['Salary Estimate'].apply(lambda x: x.split('-')[1])
df
```

```python
df['Min Salary'] = df['Min Salary'].apply(lambda x: int(x))
df['Max Salary'] = df['Max Salary'].apply(lambda x: int(x))
df
```

```python
df['Avg Salary'] = (df['Max Salary'] + df['Min Salary'])/2
df
```

```python
df['Company Name'] = df['Company Name'].apply(lambda x: str(x).split('\n')[0])
df
```

```python
df['State'] = df['Location'].apply(lambda x: x.split(',')[1])
df
```

```python
def find_lang(jd):
  jd = str(jd)
  prog_lang = ['python','r-programming','sql','tableau','powerbi']
  jd = jd.lower()
  ans = ""
  for p in prog_lang:
    if p in jd:
      ans+=p
      ans+=','
  ans = ans[:-1]
  return ans
```

```python
df['Programming Lang'] = df['Job Description'].apply(lambda x: find_lang(x))
df
```

```python
df['Age'] = df['Founded'].apply(lambda x: x if x < 0 else 2024 - x)
df
```

```python
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df
```

```python
df.columns
```

```python
df_out = df.drop('Unnamed: 0', axis=1)
df_out.columns
```

```python
df_out.to_csv('DataAnalyst_Cleaned.csv', index = False)
```
